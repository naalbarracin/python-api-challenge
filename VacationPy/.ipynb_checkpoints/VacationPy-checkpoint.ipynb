### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Create DataFrame called hotel_df to store hotel names along with city, country and coordinates
hotel = new_citys[["name", "sys.country", "coord.lat", "coord.lon"]].copy()
hotel["Hotel_name"] = ""
hotel.rename(columns={'sys.country': 'country'}, inplace=True)
hotel
```

```python
# Set parameters to search for a hotel
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key
}

# Iterate through 
for index, row in hotel.iterrows():
    # get lat, lng from df
    lat = row["coord.lat"]
    lng = row["coord.lon"]
    params["location"] = f"{lat},{lng}"
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    result = requests.get(url, params=params)
    result = result.json()
    
    # Grab the first hotel from the results and store the name
    try:
        hotel.loc[index, "Hotel_name"] = result["results"][0]["name"]
    except:
        print("Error")

hotel
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel_name}</dd>
<dt>City</dt><dd>{name}</dd>
<dt>Country</dt><dd>{country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel.iterrows()]
locations = hotel[["coord.lat", "coord.lon"]]
```

```python
# Add marker layer ontop of heat map
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)

# Display figure
fig
```
